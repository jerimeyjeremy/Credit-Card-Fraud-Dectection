```python
'''Seperate data into dependent and independent features'''
X = df.drop(['id', 'Class'], axis = 1)
Y = df.Class

#Print feature shapes
print(f"Features Shape: {X.shape}")
print(f"Target Shape: {Y.shape}")
```

```python
from sklearn.preprocessing import StandardScaler

#Scale 'Amount'
stn_scaler = StandardScaler()

X_scaled = stn_scaler.fit_transform(X)

X = pd.DataFrame(X_scaled,columns=X.columns)
```

```python
X.head(5)
```

```python
from sklearn.model_selection import train_test_split


train_x, test_x, train_y, test_y = train_test_split(X, Y, test_size = 0.3, random_state = 123)
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# Separate the data into two groups based on the target variable
group_0 = df[df["Class"] == 0]
group_1 = df[df["Class"] == 1]

# Initialize an empty dictionary to store p-values
p_values = {}

# Loop through each feature and calculate the t-test p-value
for column in df.columns[:-1]:  # Exclude the target variable
    t_stat, p_value = ttest_ind(group_0[column], group_1[column], equal_var=False)
    p_values[column] = p_value

# Convert the dictionary to a pandas Series for easier analysis
p_values_series = pd.Series(p_values, name='p-value').sort_values(ascending=True)

# Display the p-values
# Filter and print p-values greater than 0.05
significant_p_values = p_values_series[p_values_series > 0.05]
print(significant_p_values)
```

```python

import tensorflow as tf 
from tensorflow import keras
```

```python
from keras.models import Sequential
from keras import layers



model = Sequential()
model.add(layers.Dense(10, activation='relu', input_shape = (29,)))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', 
              optimizer='adam', 
               metrics=['accuracy'])

model.summary()
```

```python

history = model.fit(train_x, train_y,
                     epochs=50,
                     verbose=False,
                     validation_data=(test_x, test_y),
                     batch_size=10)
```

```python
history.params
```

```python
history.history
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

def plot_history(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    x = range(1, len(acc) + 1)

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(x, acc, 'b', label='Training acc')
    plt.plot(x, val_acc, 'r', label='Validation acc')
    plt.title('Training and validation accuracy')
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(x, loss, 'b', label='Training loss')
    plt.plot(x, val_loss, 'r', label='Validation loss')
    plt.title('Training and validation loss')
    plt.legend()
```

```python
plot_history(history)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.DataFrame(history.history).plot(figsize=(8, 5))
plt.grid(True)
plt.gca().set_ylim(0, 1) # set the vertical range to [0-1]
plt.show()
```

```python
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score

# Assuming you already have your model and predictions
predictions = model.predict(test_x)
predictions = np.round(predictions)

# Calculate Precision, Recall, and F1-score
precision = precision_score(test_y, predictions)
recall = recall_score(test_y, predictions)
f1 = f1_score(test_y, predictions)

print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1-score: {f1:.4f}')
```

```python
loss, accuracy = model.evaluate(train_x, train_y, verbose=False)
print("Training Accuracy: {:.4f}".format(accuracy))
loss, accuracy = model.evaluate(test_x, test_y, verbose=False)
print("Testing Accuracy:  {:.4f}".format(accuracy))
```
